```python
import pandas as pd
```

```python
audit = pd.read_csv("../datasets/비상장_현금흐름.csv", index_col=0)
audit
```

```python
def audit_life_cycle(data):
    data["GAAP 기업수명주기"] = ""
    data["IFRS(개별) 기업수명주기"] = ""
    data["IFRS(연결) 기업수명주기"] = ""

    ### GAAP 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[3]][i] < 0)
            & (data[data.columns[4]][i] < 0)
            & (data[data.columns[5]][i] > 0)
        ):
            data.at[i, "GAAP 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[3]][i] > 0)
            & (data[data.columns[4]][i] < 0)
            & (data[data.columns[5]][i] > 0)
        ):
            data.at[i, "GAAP 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[3]][i] > 0)
            & (data[data.columns[4]][i] < 0)
            & (data[data.columns[5]][i] < 0)
        ):
            data.at[i, "GAAP 기업수명주기"] = "성숙기"
        elif (data[data.columns[3]][i] < 0) & (data[data.columns[4]][i] > 0):
            data.at[i, "GAAP 기업수명주기"] = "쇠퇴기"

    ### IFRS(개별) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[6]][i] < 0)
            & (data[data.columns[7]][i] < 0)
            & (data[data.columns[8]][i] > 0)
        ):
            data.at[i, "IFRS(개별) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[6]][i] > 0)
            & (data[data.columns[7]][i] < 0)
            & (data[data.columns[8]][i] > 0)
        ):
            data.at[i, "IFRS(개별) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[6]][i] > 0)
            & (data[data.columns[7]][i] < 0)
            & (data[data.columns[8]][i] < 0)
        ):
            data.at[i, "IFRS(개별) 기업수명주기"] = "성숙기"
        elif (data[data.columns[6]][i] < 0) & (data[data.columns[7]][i] > 0):
            data.at[i, "IFRS(개별) 기업수명주기"] = "쇠퇴기"

    ### IFRS(연결) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[9]][i] < 0)
            & (data[data.columns[10]][i] < 0)
            & (data[data.columns[11]][i] > 0)
        ):
            data.at[i, "IFRS(연결) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[9]][i] > 0)
            & (data[data.columns[10]][i] < 0)
            & (data[data.columns[11]][i] > 0)
        ):
            data.at[i, "IFRS(연결) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[9]][i] > 0)
            & (data[data.columns[10]][i] < 0)
            & (data[data.columns[11]][i] < 0)
        ):
            data.at[i, "IFRS(연결) 기업수명주기"] = "성숙기"
        elif (data[data.columns[9]][i] < 0) & (data[data.columns[10]][i] > 0):
            data.at[i, "IFRS(연결) 기업수명주기"] = "쇠퇴기"
    return data
```

```python
audit_lc = audit_life_cycle(audit)
audit_lc
```

```python
audit_label = audit_lc[['회사명', '거래소코드', '회계년도', 'GAAP 기업수명주기', 'IFRS(개별) 기업수명주기', 'IFRS(연결) 기업수명주기']]
```

```python
audit_label.to_csv("../datasets/Labeling/비상장_수명주기처리전.csv")
```

### 수명주기 처리 전 데이터프레임 가져오기

```python
audit = pd.read_csv("../datasets/Labeling/비상장_수명주기처리전.csv", index_col=0)
```

```python
audit["회계년도"] = pd.to_datetime(audit["회계년도"]).dt.year
```

```python
# 결측치가 있는 행 제거
df_cleaned_audit = audit.dropna(
    subset=[
        "GAAP 기업수명주기",
        "IFRS(개별) 기업수명주기",
        "IFRS(연결) 기업수명주기",
    ],
    how="all",
)

print(len(df_cleaned_audit["회사명"].unique()))
print(len(df_cleaned_audit["거래소코드"].unique()))
```

```python
df_cleaned_audit = df_cleaned_audit.reset_index(drop=True)
```

```python
a1 = df_cleaned_audit.columns[3]
a2 = df_cleaned_audit.columns[4]
a3 = df_cleaned_audit.columns[5]
```

```python
print(df_cleaned_audit.isnull().sum())
```

### audit 기업수명주기 결측치 채우기
#### 비상장 외감 기업( IFRS연결 → IFRS개별 → GAAP개별)

```python
nan_list = []
a2_list = []
a3_list = []

for i in range(len(df_cleaned_audit)):
    # GAAP 결측치
    if pd.isna(df_cleaned_audit[a1][i]):
        nan_list.append(df_cleaned_audit["거래소코드"][i])
        if pd.notna(df_cleaned_audit[a2][i]):
            a2_list.append(df_cleaned_audit["거래소코드"][i])
        if pd.notna(df_cleaned_audit[a3][i]):
            a3_list.append(df_cleaned_audit["거래소코드"][i])

print(nan_list)
print(len(nan_list))

print(a2_list)
print(len(a2_list))

print(a3_list)
print(len(a3_list))
```

```python
df_cleaned_audit["GAAP 기업수명주기"] = df_cleaned_audit["GAAP 기업수명주기"].fillna(
    df_cleaned_audit["IFRS(연결) 기업수명주기"]
)
```

```python
print(df_cleaned_audit.isnull().sum())
```

```python
audit_life_cycle = df_cleaned_audit[["회사명", "거래소코드", "회계년도", "GAAP 기업수명주기"]]
```

```python
audit_life_cycle.columns = ["회사명", "거래소코드", "회계년도", "기업수명주기"]
```

```python
audit_life_cycle.to_csv("../datasets/Labeling/비상장_수명주기.csv")
```
