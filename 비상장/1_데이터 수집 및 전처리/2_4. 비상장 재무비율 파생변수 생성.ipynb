- 영업레버리지 = 매출총이익 / 영업이익

```python
# 비상장 국제회계기준 연결 영업이익
비상국제연결영업이익 = 비상재무['[U01B420000000]* (정상)영업손익(보고서기재)(IFRS연결)(천원)'].fillna(비상재무['[U01B430000000]* (정상)영업손익(계산수치)(IFRS연결)(천원)'])

# 비상장 국제회계기준 연결 영업레버리지
비상국제연결영업레버리지 = 비상재무['[U01B201014400]매출총이익(손실)(IFRS연결)(천원)'] / 비상국제연결영업이익

# 비상장 국제회계기준 개별 영업이익
비상국제개별영업이익 = 비상재무['[U01B420000000]* (정상)영업손익(보고서기재)(IFRS)(천원)'].fillna(비상재무['[U01B430000000]* (정상)영업손익(계산수치)(IFRS)(천원)'])

# 비상장 국제회계기준 개별 영업레버리지
비상국제개별영업레버리지 = 비상재무['[U01B201014400]매출총이익(손실)(IFRS)(천원)'] / 비상국제개별영업이익

# 비상장 일반회계기준 연결은 없음
# 비상장 일반회계기준 개별 영업이익
비상일반개별영업이익 = 비상재무['[U01230000]영업이익(손실)(천원)']

# 비상장 일반회계기준 개별 영업레버리지
비상일반개별영업레버리지 = 비상재무['[U01221340]매출총이익(손실)(천원)'] / 비상일반개별영업이익

# 우선순위로 fillna
비상영업레버리지 = 비상국제연결영업레버리지.fillna(비상국제개별영업레버리지)
비상영업레버리지 = 비상영업레버리지.fillna(비상일반개별영업레버리지)

# 새 컬럼에 입력
비상재무['영업레버리지'] = 비상영업레버리지
```

- 재무레버리지 = 영업이익 / 세전이익

```python
# 비상장 국제회계기준 연결 재무레버리지
비상국제연결재무레버리지 = 비상국제연결영업이익 / 비상재무['[U01B700000000]법인세비용차감전순이익(손실)(IFRS연결)(천원)']

# 비상장 국제회계기준 개별 재무레버리지
비상국제개별재무레버리지 = 비상국제개별영업이익 / 비상재무['[U01B700000000]법인세비용차감전순이익(손실)(IFRS)(천원)']

# 비상장 일반회계기준 개별 재무레버리지
비상일반개별재무레버리지 = 비상일반개별영업이익 / 비상재무['[U01290000]법인세비용차감전(계속사업)손익(천원)']

# 우선순위로 fillna
비상재무레버리지 = 비상국제연결재무레버리지.fillna(비상국제개별재무레버리지)
비상재무레버리지 = 비상재무레버리지.fillna(비상일반개별재무레버리지)

# 새 컬럼에 입력
비상재무['재무레버리지'] = 비상재무레버리지
```
