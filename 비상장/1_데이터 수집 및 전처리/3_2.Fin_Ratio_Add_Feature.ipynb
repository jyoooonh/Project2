```python
import pandas as pd
```

##### 재무비율 불러오기

```python
fin = pd.read_csv("../1.데이터셋/비상장_재무비율.csv")
```

##### 이모 피처 불러오기

```python
joo = pd.read_csv("../1.데이터셋/피처생성/비상장_주피처.csv", index_col=0)
joo["회계년도"] = pd.to_datetime(joo["회계년도"])
joo["회계년도"] = pd.to_datetime(joo["회계년도"]).dt.year
joo
```

##### 추가피처 가져오기

```python
croe = pd.read_csv('../1.데이터셋/피처생성/비상장추가피처.csv', index_col=0)
croe
```

##### 상장현금흐름 파생변수 가져오기

```python
cf = pd.read_csv("../1.데이터셋/피처생성/비상장현금흐름파생변수.csv", index_col=0)
cf["회계년도"] = pd.to_datetime(cf["회계년도"])
cf["회계년도"] = pd.to_datetime(cf["회계년도"]).dt.year
```

##### 5개 데이터프레임 merge

```python
df = pd.merge(fin, joo, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df = pd.merge(df, croe, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df = pd.merge(df, cf, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df["설립일"] = pd.to_datetime(df["설립일"]).dt.year
df["영업년수"] = 2024 - df["설립일"]
```

```python
df.to_csv('../1.데이터셋/Feature/1.재무비율.csv', index=False)
```
