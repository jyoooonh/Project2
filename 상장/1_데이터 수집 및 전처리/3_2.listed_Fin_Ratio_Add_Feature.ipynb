```python
import pandas as pd
```

##### 재무비율 불러오기

```python
fin = pd.read_csv("../datasets/상장_재무비율.csv")
fin["회계년도"] = pd.to_datetime(fin["회계년도"], format="%y-%b")
fin["회계년도"] = pd.to_datetime(fin["회계년도"]).dt.year
```

##### 시장지표 불러오기

```python
market = pd.read_csv("../datasets/피처생성/상장_시장지표.csv", index_col=0)
market["회계년도"] = pd.to_datetime(market["회계년도"])
market["회계년도"] = pd.to_datetime(market["회계년도"]).dt.year
```

##### 이모 피처 불러오기

```python
joo = pd.read_csv("../datasets/피처생성/상장_주피처.csv", index_col=0)
joo["회계년도"] = pd.to_datetime(joo["회계년도"])
joo["회계년도"] = pd.to_datetime(joo["회계년도"]).dt.year
```

##### 추가피처 가져오기

```python
croe = pd.read_csv('../datasets/피처생성/상장추가피처.csv', index_col=0)
```

##### 상장현금흐름 파생변수 가져오기

```python
cf = pd.read_csv("../datasets/피처생성/상장현금흐름파생변수.csv", index_col=0)
cf["회계년도"] = pd.to_datetime(cf["회계년도"])
cf["회계년도"] = pd.to_datetime(cf["회계년도"]).dt.year
```

##### 5개 데이터프레임 merge

```python
df = pd.merge(fin, market, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df = pd.merge(df, joo, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df = pd.merge(df, croe, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df = pd.merge(df, cf, how="left", on=["회사명", "거래소코드", "회계년도"])
df
```

```python
df.to_csv('../datasets/Feature/1.재무비율.csv', index=False)
```
