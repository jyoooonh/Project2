```python
def ko_life_cycle(data):
    ### IFRS 기업 수명 주기
    data["GAAP(개별) 기업수명주기"] = ""
    data["GAAP(연결) 기업수명주기"] = ""
    data["IFRS(개별) 기업수명주기"] = ""
    data["IFRS(개별)(비교) 기업수명주기"] = ""
    data["IFRS(연결) 기업수명주기"] = ""
    data["IFRS(연결)(비교) 기업수명주기"] = ""

    ### GAAP(개별) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[3]][i] < 0)
            & (data[data.columns[4]][i] < 0)
            & (data[data.columns[5]][i] > 0)
        ):
            data.at[i, "GAAP(개별) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[3]][i] > 0)
            & (data[data.columns[4]][i] < 0)
            & (data[data.columns[5]][i] > 0)
        ):
            data.at[i, "GAAP(개별) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[3]][i] > 0)
            & (data[data.columns[4]][i] < 0)
            & (data[data.columns[5]][i] < 0)
        ):
            data.at[i, "GAAP(개별) 기업수명주기"] = "성숙기"
        elif (data[data.columns[3]][i] < 0) & (data[data.columns[4]][i] > 0):
            data.at[i, "GAAP(개별) 기업수명주기"] = "쇠퇴기"

    ### GAAP(연결) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[6]][i] < 0)
            & (data[data.columns[7]][i] < 0)
            & (data[data.columns[8]][i] > 0)
        ):
            data.at[i, "GAAP(연결) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[6]][i] > 0)
            & (data[data.columns[7]][i] < 0)
            & (data[data.columns[8]][i] > 0)
        ):
            data.at[i, "GAAP(연결) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[6]][i] > 0)
            & (data[data.columns[7]][i] < 0)
            & (data[data.columns[8]][i] < 0)
        ):
            data.at[i, "GAAP(연결) 기업수명주기"] = "성숙기"
        elif (data[data.columns[6]][i] < 0) & (data[data.columns[7]][i] > 0):
            data.at[i, "GAAP(연결) 기업수명주기"] = "쇠퇴기"

    ### IFRS(개별) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[9]][i] < 0)
            & (data[data.columns[10]][i] < 0)
            & (data[data.columns[11]][i] > 0)
        ):
            data.at[i, "IFRS(개별) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[9]][i] > 0)
            & (data[data.columns[10]][i] < 0)
            & (data[data.columns[11]][i] > 0)
        ):
            data.at[i, "IFRS(개별) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[9]][i] > 0)
            & (data[data.columns[10]][i] < 0)
            & (data[data.columns[11]][i] < 0)
        ):
            data.at[i, "IFRS(개별) 기업수명주기"] = "성숙기"
        elif (data[data.columns[9]][i] < 0) & (data[data.columns[10]][i] > 0):
            data.at[i, "IFRS(개별) 기업수명주기"] = "쇠퇴기"

    ### IFRS(개별)(비교) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[12]][i] < 0)
            & (data[data.columns[13]][i] < 0)
            & (data[data.columns[14]][i] > 0)
        ):
            data.at[i, "IFRS(개별)(비교) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[12]][i] > 0)
            & (data[data.columns[13]][i] < 0)
            & (data[data.columns[14]][i] > 0)
        ):
            data.at[i, "IFRS(개별)(비교) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[12]][i] > 0)
            & (data[data.columns[13]][i] < 0)
            & (data[data.columns[14]][i] < 0)
        ):
            data.at[i, "IFRS(개별)(비교) 기업수명주기"] = "성숙기"
        elif (data[data.columns[12]][i] < 0) & (data[data.columns[13]][i] > 0):
            data.at[i, "IFRS(개별)(비교) 기업수명주기"] = "쇠퇴기"

    ### IFRS(연결) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[15]][i] < 0)
            & (data[data.columns[16]][i] < 0)
            & (data[data.columns[17]][i] > 0)
        ):
            data.at[i, "IFRS(연결) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[15]][i] > 0)
            & (data[data.columns[16]][i] < 0)
            & (data[data.columns[17]][i] > 0)
        ):
            data.at[i, "IFRS(연결) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[15]][i] > 0)
            & (data[data.columns[16]][i] < 0)
            & (data[data.columns[17]][i] < 0)
        ):
            data.at[i, "IFRS(연결) 기업수명주기"] = "성숙기"
        elif (data[data.columns[15]][i] < 0) & (data[data.columns[16]][i] > 0):
            data.at[i, "IFRS(연결) 기업수명주기"] = "쇠퇴기"

    ### IFRS(연결)(비교) 기업 수명 주기
    for i in range(len(data)):
        if (
            (data[data.columns[18]][i] < 0)
            & (data[data.columns[19]][i] < 0)
            & (data[data.columns[20]][i] > 0)
        ):
            data.at[i, "IFRS(연결)(비교) 기업수명주기"] = "도입기"
        elif (
            (data[data.columns[18]][i] > 0)
            & (data[data.columns[19]][i] < 0)
            & (data[data.columns[20]][i] > 0)
        ):
            data.at[i, "IFRS(연결)(비교) 기업수명주기"] = "성장기"
        elif (
            (data[data.columns[18]][i] > 0)
            & (data[data.columns[19]][i] < 0)
            & (data[data.columns[20]][i] < 0)
        ):
            data.at[i, "IFRS(연결)(비교) 기업수명주기"] = "성숙기"
        elif (data[data.columns[18]][i] < 0) & (data[data.columns[19]][i] > 0):
            data.at[i, "IFRS(연결)(비교) 기업수명주기"] = "쇠퇴기"
    return data
```

```python
kskqkn["회계년도"] = pd.to_datetime(kskqkn["회계년도"]).dt.year
```

```python
print(len(kskqkn["회사명"].unique()))
print(len(kskqkn["거래소코드"].unique()))
```

```python
# 결측치가 있는 행 제거
df_cleaned_kskqkn = kskqkn.dropna(
    subset=[
        "GAAP(개별) 기업수명주기",
        "GAAP(연결) 기업수명주기",
        "IFRS(개별) 기업수명주기",
        "IFRS(개별)(비교) 기업수명주기",
        "IFRS(연결) 기업수명주기",
        "IFRS(연결)(비교) 기업수명주기",
    ],
    how="all",
)

print(len(df_cleaned_kskqkn["회사명"].unique()))
print(len(df_cleaned_kskqkn['거래소코드'].unique()))
```

```python
df_cleaned_kskqkn = df_cleaned_kskqkn.reset_index(drop=True)
```

```python
k1 = df_cleaned_kskqkn.columns[3]
k2 = df_cleaned_kskqkn.columns[4]
k3 = df_cleaned_kskqkn.columns[5]
k4 = df_cleaned_kskqkn.columns[6]
k5 = df_cleaned_kskqkn.columns[7]
k6 = df_cleaned_kskqkn.columns[8]
```

```python
print(df_cleaned_kskqkn.isnull().sum())
```

```python
nan_list = []

# IFRS 연결과 IFRS 개별이 다른 것
diff1_list = []
# IFRS 연결과 GAAP 연결이 다른 것
diff2_list = []

for i in range(len(df_cleaned_kskqkn)):
    if pd.isna(df_cleaned_kskqkn[k5][i]):
        nan_list.append(df_cleaned_kskqkn["거래소코드"][i])
    if pd.notna(df_cleaned_kskqkn[k5][i]) & pd.notna(df_cleaned_kskqkn[k3][i]):
        if df_cleaned_kskqkn[k5][i] != df_cleaned_kskqkn[k3][i]:
            diff1_list.append(df_cleaned_kskqkn["거래소코드"][i])
    elif pd.notna(df_cleaned_kskqkn[k5][i]) & pd.notna(df_cleaned_kskqkn[k2][i]):
        if df_cleaned_kskqkn[k5][i] != df_cleaned_kskqkn[k2][i]:
            diff2_list.append(df_cleaned_kskqkn["거래소코드"][i])


print(nan_list)
print(len(nan_list))

print(diff1_list)
print(len(diff1_list))

print(diff2_list)
print(len(diff2_list))

print(len(set(diff1_list)))
```

#### 상장 기업(IFRS연결 →  IFRS개별 → GAAP연결 → GAAP개별 → IFRS(개별)(비교) → IFRS(연결)(비교))

```python
df_cleaned_kskqkn["IFRS(연결) 기업수명주기"] = df_cleaned_kskqkn["IFRS(연결) 기업수명주기"].fillna(
    df_cleaned_kskqkn["IFRS(개별) 기업수명주기"]
)
print(df_cleaned_kskqkn.isnull().sum())
```

```python
df_cleaned_kskqkn["IFRS(연결) 기업수명주기"] = df_cleaned_kskqkn["IFRS(연결) 기업수명주기"].fillna(
    df_cleaned_kskqkn["GAAP(연결) 기업수명주기"]
)
print(df_cleaned_kskqkn.isnull().sum())
```

```python
df_cleaned_kskqkn["IFRS(연결) 기업수명주기"] = df_cleaned_kskqkn["IFRS(연결) 기업수명주기"].fillna(
    df_cleaned_kskqkn["GAAP(개별) 기업수명주기"]
)
print(df_cleaned_kskqkn.isnull().sum())
```

```python
df_cleaned_kskqkn["IFRS(연결) 기업수명주기"] = df_cleaned_kskqkn["IFRS(연결) 기업수명주기"].fillna(
    df_cleaned_kskqkn["IFRS(개별)(비교) 기업수명주기"]
)
print(df_cleaned_kskqkn.isnull().sum())
```

```python
df_cleaned_kskqkn["IFRS(연결) 기업수명주기"] = df_cleaned_kskqkn["IFRS(연결) 기업수명주기"].fillna(
    df_cleaned_kskqkn["IFRS(연결)(비교) 기업수명주기"]
)
print(df_cleaned_kskqkn.isnull().sum())
```
