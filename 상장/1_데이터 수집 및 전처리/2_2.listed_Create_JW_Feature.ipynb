상장영업현금

```python
상장영업현금흐름 = oncash['[U01D100000000]영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'].fillna(oncash['[U01D100000000]영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)'])
상장영업현금흐름 = 상장영업현금흐름.fillna(oncash['[U01410000]영업활동으로 인한 현금흐름(*)(연결)(천원)'])
상장영업현금흐름 = 상장영업현금흐름.fillna(oncash['[U01410000]영업활동으로 인한 현금흐름(*)(천원)'])
oncash['상장영업현금'] = 상장영업현금흐름
oncash.drop(
    labels=[
        '[U01D100000000]영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)',
        '[U01D100000000]영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)',
        '[U01410000]영업활동으로 인한 현금흐름(*)(연결)(천원)',
        '[U01410000]영업활동으로 인한 현금흐름(*)(천원)',
        '[U01D100000000]영업활동으로 인한 현금흐름(간접법)(*)(비교)(IFRS연결)(천원)',
        '[U01D100000000]영업활동으로 인한 현금흐름(간접법)(*)(비교)(IFRS)(천원)'
    ],
    axis=1,
    inplace=True
)
```

상장재무현금

# 파생변수

- OMS-CF : 영업현금흐름 - 재무현금흐름

```python
# oncash['OMS(천원)'] = oncash['상장영업현금'] - oncash['상장재무현금']
```

- ODI-CF : 영업현금흐름 / 투자현금흐름

```python
oncash['영업투자현금비율'] = oncash['상장영업현금'] / oncash['상장투자현금']
```

- ODS-CF : 영업현금흐름 / 재무현금흐름

```python
oncash['영업재무현금비율'] = oncash['상장영업현금'] / oncash['상장재무현금']
```
